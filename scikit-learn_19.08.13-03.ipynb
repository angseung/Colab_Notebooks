```python
import pandas as pd
#!ls '/content/drive/'
#!ls "/gdrive/My Drive/Colab Notebooks/movie_data.csv"
df = pd.read_csv('https://drive.google.com/file/d/17pSHp8mb_RT4c-W7bZB1NoNL18S4t36-/view?usp=sharing', encoding='utf-8')
df.head(3)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

count = CountVectorizer(stop_words='english',
                        max_df=.1,
                        max_features=5000)
X = count.fit_transform(df['review'].values) #shape: (50000,5000)
```

```python
#실행 엄청 오래 걸림!
from sklearn.decomposition import LatentDirichletAllocation

lda = LatentDirichletAllocation(n_components=10,
                                random_state=123,
                                learning_method='batch')
X_topics = lda.fit_transform(X)
```

```python
n_top_words = 10
feature_names = count.get_feature_names()

for topic_idx, topic in enumerate(lda.components_):
    print("토픽 %d:" % (topic_idx + 1))
    print(" ".join([feature_names[i]
                    for i in topic.argsort()\
                        [:-n_top_words - 1:-1]]))
```

```python
lda.components_[4, :].argsort()[:-6:-1]
```

```python
for i in lda.components_[0, :].argsort()[:-6:-1]:
    print(feature_names[i])
```

```python
comedy = X_topics[:, 6].argsort()[::-1] #어떤 topic에 속할 확률이 가장 큰 5개의 index가 반환된다.

for iter_idx, movie_idx in enumerate(comedy[:3]):
    print('\n코미디 영화 #%d:' % (iter_idx + 1))
    print(df['review'][movie_idx][:1000], '...')
    print(movie_idx)
```
